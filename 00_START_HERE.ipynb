## 7. Suggested Workflow

A simple pattern for the first few weeks of a project is:

1. Run `make init` once after pulling new environment changes.
2. Write or edit code in your package, not only in notebooks.
3. Run `make test` after each small change.
4. Add `make format` or `make lint` if style and bug-catching feedback would help.
5. Build docs with `make docs` when your docstrings become useful to readers.

If your team is using GitHub pull requests, `make check-full` is a good pre-share command once you are comfortable with the optional tools.
