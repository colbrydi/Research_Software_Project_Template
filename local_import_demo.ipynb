## How this works

If the notebook is opened from the repository root, `mypackage` can be imported directly.
The next cell also appends the current working directory to `sys.path` for clarity.

```python
from pathlib import Path
import sys

repo_root = Path.cwd()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

print(f'Repository root in sys.path: {repo_root in map(Path, map(Path, sys.path))}')
```

```python
from mypackage import power_self

value = power_self(3)
print(f'power_self(3) = {value}')
```
